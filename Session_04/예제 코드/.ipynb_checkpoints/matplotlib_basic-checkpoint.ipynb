# Matplotlib Basic

```python
from matplotlib import pyplot as plt #matplotlib 모듈 안에 세부 모듈 pyplot이 있어서 from으로 시작하는 구문이 한번 더 있는 것
```

### Line Chart

```python
years = list(range(2007, 2019))
wage = [3480, 3770, 4000, 4110, 4320, 4580, 4860, 5120, 5580, 6030, 6470, 7530]
```

```python
plt.plot(years, wage) # plt.plot(x, y): (x, y) 쌍을 연결한 line chart를 반환 .
plt.show() # 시각화 세팅을 마친 뒤에는 항상 show()를 해주어야 최종적으로 출력됨 & 시각화 설정 초기화됨.
```

```python
plt.plot(years, wage, color = 'green', marker = 'x', linestyle = 'dashed', linewidth = 2, label = 'wage')
# 색상, 점 모양, 선 모양, 선 굵기 등을 설정한 line chart를 반환.
plt.title('Minimum Wage') # 타이틀 설정
plt.xlabel('year') # x 라벨 설정
plt.ylabel('min-wage') # y 라벨 설정
plt.grid(True) # grid를 그림
plt.legend() # 범례를 표시- 이걸 하려면 plot 함수에서 label 설정을 반드시 해주어야.

plt.show()
```

```python
plt.plot(years, wage, color = 'green', marker = 'x', linestyle = 'dashed', linewidth = 2, label = 'wage')
plt.plot(years, sorted(wage, reverse=True), color = 'blue', marker = 'o', linestyle = 'solid', linewidth = 2, label = 'rev_wage')
plt.title('Minimum Wage')
plt.xlabel('year')
plt.ylabel('min-wage')
plt.grid(True)
plt.legend()

plt.show() # 두 개 이상의 선 그래프를 동시에 도시할 수도 있다.
```

### Bar Chart

```python
movies = ['Inception', 'Harry Potter', 'Lord of Rings', 'Get Out', 'Thor']
grades = [9.2, 5.1, 3.7, 0, 10]
xs = [i+0.1 for i, _ in enumerate(movies)]
```

```python
plt.bar(xs, grades) # Bar Chart를 반환 - xs는 막대의 위치, grades는 막대의 높이.
plt.xticks(xs, movies) # 각 xs의 위치마다 movies의 원소로 라벨링을 해줌.

plt.show()
```

### Histogram

```python
import pandas as pd
df = pd.read_csv('C:\네이버.csv', engine='python', index_col=0)
df.head()
```

```python
plt.hist(df['시가']) # hist() 함수는 한 개 데이터만을 인자로 받음.
plt.show()
```

```python
plt.hist(df['시가'], bins=20) # bins 설정 시 구간의 개수를 설정할 수 있음.
plt.show()
```

### Scatter Plot

```python
plt.scatter(df['시가'], df['종가']) # scatter() 함수는 두 개의 인자가 필요함.
plt.show()
```

```python
plt.scatter(df['시가'][:10], df['종가'][:10])
for label, x, y in zip(df['일자'][:10], df['시가'][:10], df['종가'][:10]):
    plt.annotate(label, xy=(x,y)) # annotate()를 사용 시 지정된 (x, y) 좌표에 label을 붙일 수 있음.

plt.show()
```

### Subplot

```python
plt.figure(figsize=(20,10)) # figure() 함수를 이용하면 차트에 대한 각종 설정 가능.
                            # 여기서는 figsize parameter에 너비, 높이 튜플을 할당해 차트의 크기 조정.

plt.subplot(121) # 1*2로 쪼갠 뒤, 그 중 첫 번째에 차트를 그리겠다는 선언.
plt.hist(df['시가'], bins=20)

plt.subplot(122) # 두 번째에 차트를 그리겠다는 선언.
plt.scatter(df['시가'], df['종가'])

plt.show()
```

### 응용: Word Cloud

```python
data = [("big data", 100, 15), ("Hadoop", 95, 25), ("Python", 75, 50), ("R", 50, 40), ("machine learning", 80, 20), ("statistics", 20, 60), ("data science", 60, 70), ("analytics", 90, 3), ("team player", 85, 85), ("dynamic", 2, 90), ("synergies", 70, 0), ("actionable insights", 40, 30), ("think out of the box", 45, 10), ("self-starter", 30, 50), ("customer focus", 65, 15), ("thought leadership", 35, 35)]

text_size = lambda x: 8 + x / 200 * 20

for word, job_popularity, resume_popularity in data:
    plt.text(job_popularity, resume_popularity, word, ha='center', va='center',size=text_size(job_popularity + resume_popularity))

plt.xlabel("Popularity on Job Postings")
plt.ylabel("Popularity on Resumes")`
```
