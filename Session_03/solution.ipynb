## Quest solution

```python
import numpy as np
import scipy.optimize as op
import matplotlib.pyplot as plt
```

```python
# 2차원 cost function
def f2(x): 
    return (1-x[0])**2 + 100*(x[1]-x[0]**2)**2

initial_guess = (1,1) # or [1,1]처럼 리스트 형태!
# initial guess는 임의의 값. 다만 2차원 cost function의 식은 간단해서 최적 파라미터값 예상이 가능하기 때문에 (1,1)을 initial guess로 잡은 것

result = op.minimize(f2, initial_guess)
result

# 퀘스트 끝..!
```

```python
print('fun:', result.fun,': min f(x,y), 함수의 최솟값\n', 'x:', result.x, ': 최적 파라미터', '\n즉 x=1, y=1에서 2차원 cost function이 최소화')
# initial guess값에 따라 조금씩 결과가 달라질 수 있지만 모두 1에 거의 가까운 값이 나올 것
```
