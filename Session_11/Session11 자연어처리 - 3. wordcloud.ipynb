```python
from wordcloud import WordCloud, STOPWORDS

import numpy as np
from PIL import Image
```

```python
text = open('robinson_crusoe.txt',encoding='utf-8').read()
```

```python
rc_mask = np.array(Image.open('img/island.png'))
```

```python
stopwords =set(STOPWORDS)
```

```python
#said란 단어가 많이 등장하는데, stopwords에는 포함이 안되어있으므로 추가해줍니다.
stopwords.add("said")
```

```python
import matplotlib.pyplot as plt
path = "c:/Windows/Fonts/malgun.ttf"
```

```python
plt.figure(figsize=(10,10))
plt.imshow(rc_mask, cmap=plt.cm.gray, interpolation = 'bilinear')
plt.axis('off')
plt.show()
```

```python
wc = WordCloud(background_color='white', max_words=2000, mask=rc_mask, colormap='Greens', stopwords = stopwords)
wc = wc.generate(text)
```

```python
plt.figure(figsize=(15,15))
plt.imshow(wc.recolor(random_state = 5), interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
from wordcloud import WordCloud, STOPWORDS
from PIL import Image
from wordcloud import ImageColorGenerator

text = open('robinson_crusoe.txt',encoding='utf-8').read()

colored_gh = np.array(Image.open("gh_colored_2.png"))
stopwords = set(STOPWORDS)
stopwords.add("said")

wc = WordCloud(background_color="white", max_words=800, mask=colored_gh,
               stopwords=stopwords, max_font_size=40, random_state=-9)
# generate word cloud
wc.generate(text)

# create coloring from image
image_colors = ImageColorGenerator(colored_gh)

# show
plt.figure(figsize=(15,15))
# recolor wordcloud and show
# we could also give color_func=image_colors directly in the constructor
plt.imshow(wc.recolor(color_func=image_colors), interpolation="bilinear")
plt.axis('off')
plt.show()
```

```python
#너무 빽빽한것 같으니 max_words를 400으로 줄여봅시다
wc = WordCloud(background_color="white",max_words=400, mask=colored_gh,
               stopwords=stopwords, max_font_size=40, random_state=None)
# generate word cloud
wc.generate(text)

# create coloring from image
image_colors = ImageColorGenerator(colored_gh)

# show
plt.figure(figsize=(15,15))
# recolor wordcloud and show
# we could also give color_func=image_colors directly in the constructor
plt.imshow(wc.recolor(color_func=image_colors), interpolation="bilinear")
plt.axis('off')
plt.show()
```
